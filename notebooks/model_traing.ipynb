```python
df['date']=pd.to_datetime(df['date'])
df.info()
```

```python
df.dropna(inplace=True) # remove missing value from the whole data set 
df.shape
```

```python
#  feature engg.  
df['total_review'] =df['drugName'] + df['condition'] +df['review']
df.head()
```

```python
df['clean_review'] = (df['total_review']
                      .str.lower()
                      .str.replace(r'\d','',regex=True)
                      .str.replace(r'[^\w\s]','',regex=True)
                      )
df['clean_review']
```

```python
from nltk.corpus import stopwords
stop_word=set(stopwords.words('english'))
df['clean_review'] = df['clean_review'].apply(lambda x:' '.join(w for w in x.split() if w not in stop_word)) 
df['clean_review']
```

```python
# add sentiment column..
df['sentiment'] = df['rating'].apply(lambda x: 'postive' if x>5 else 'negative')
df['sentiment']
```

```python
df_clean= df[['total_review','drugName','condition','sentiment','rating']]
df_clean
```
